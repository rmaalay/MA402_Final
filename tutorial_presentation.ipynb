# PETSc KSP Solver Demo

We solve Ax = b using PETSc via petsc4py.

```python
from tutorial_module import Poisson2DKSPSolver
```

```python
solver = Poisson2DKSPSolver(20, 20)
result = solver.solve()
```

```python
print("Error:", result.error_norm)
print("Iterations:", result.iterations)
```

```python
import matplotlib.pyplot as plt

grid = solver.solution_as_grid(result.x)

plt.imshow(grid)
plt.colorbar()
plt.title("Solution")
plt.show()
```
